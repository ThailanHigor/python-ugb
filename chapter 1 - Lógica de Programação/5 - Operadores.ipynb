# Operadores aritméticos


| Operador | Significado   | Exemplo                    |
|:---|:---|:---|
|     +    | Adição        | a = 3 + 2                  |
|     -    | Subtração     | b = 7 - a                  |
|     *    | Multiplicação | c = a * b                  |
|     /    | Divisão       | d = 10/3                   |
|    //    | Quociente     | e = 10//3                  |
|     %    | Resto         | f = 10%3                   |
|    **    | Exponenciação | g = 2 ** 3                 |
|    **    | Radiciação    | h = 16 ** (1/2)            |

```python
a = 3 + 2
```

```python
b = 7 - a
print(b)
```

```python
c = a * b
print(c)
```

```python
d = 10 / 3
print(d)
```

```python
e = 10 // 3
print(e)
```

```python
f = 10 % 3
print(f)
```

```python
g = 2 ** 3
print(g)
```

```python
h = 16 ** (1/2)
print(h)
```

# Operadores lógicos

| Operador | Significado   | Exemplo                    |
|:---|:---|:---|
| and | **E** lógico | a == b and b == c |
| or | **OU** lógico | a == b or b == c |
| ^ | **OU** exclusivo (XOR) | a == b ^ b == c |
| not | Negação (inverte o resultado lógico) | not(a == b and b == c) |

### Tabela verdade (AND)

| a | b | a and b |
|:--:|:--:|:--:|
| True  | True  | **True** |
| True  | False | False |
| False | True  | False |
| False | False | False |


### Tabela verdade (OR)

| a | b | a and b |
|:--:|:--:|:--:|
| True  | True  | **True** |
| True  | False | **True** |
| False | True  | **True** |
| False | False | False |


### Tabela verdade (NOT)
| a | not(a) |
|:--:|:--:|
| True  | False |
| False | **True** |

```python
# TABELA DO AND (com 2 booleanos)
print(" x     | y     | x and y ")
print("-------------------------")
print(" True  | True  |", True and True )
print(" True  | False |", True and False )
print(" False | True  |", False and True)
print(" False | False |", False and False)
```

```python
# TABELA DO OR
print(" x     | y     | x or y ")
print("-------------------------")
print(" True  | True  |",True or True )
print(" True  | False |",True or False )
print(" False | True  |",False or True)
print(" False | False |",False or False)
```

```python
# TABELA DO NOT
print(" x     | not(x) ")
print("----------------")
print(" True  |",not(True))
print(" False |",not(False))
```

# Operadores de identidade

| Operador | Significado   | Exemplo                    |
|:---|:---|:---|
| is | Verifica se duas referências apontam para o mesmo objeto | a is b |
| is not | Verifica se duas referências apontam para diferentes objetos | a is not b |


```python
a = 1000
b = 1000
c = a
print("a is b = ", a is b)
print("a is c = ", a is c)
print("b is c = ", b is c)
```
